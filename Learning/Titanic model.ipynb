```python
import re

import sys
# sys.path.append("..")
# from shad_util import print_answer

data = train_df
def clean_name(name):
    # Первое слово до запятой - фамилия
    s = re.search('^[^,]+, (.*)', name)
    if s:
        name = s.group(1)

    # Если есть скобки - то имя пассажира в них
    s = re.search('\(([^)]+)\)', name)
    if s:
        name = s.group(1)

    # Удаляем обращения
    name = re.sub('(Miss\. |Mrs\. |Ms\. )', '', name)

    # Берем первое оставшееся слово и удаляем кавычки
    name = name.split(' ')[0].replace('"', '')

    return name

names = data[data['Sex'] == 'female']['Name'].map(clean_name)
name_counts = names.value_counts()
# print(data[data['Sex'] == 'female']['Name'])
# print_answer(6, name_counts.head(1).index.values[0])
# x = train_df['Sex']
# # print(x)
# # female = train_df[['Sex', 'Name']].groupby(['Sex'])
# # print(female.head())
# # print(train_df['Name'].value_counts(sort=False))
# arr_female = []
# for item in train_df['Name']:
#     print(item)
#     if item == 'female':
#         arr_female.append(item)
# print('arr_female\n',arr_female)

ages = data['Age'].dropna()
print(ages)
print("{:0.2f} {:0.2f}".format(ages.mean(), ages.median()))
```
